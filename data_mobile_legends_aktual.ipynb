## Menambahkan Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Mengidentifikasi Variabel

```python
df_alt = pd.read_csv ('/home/adamyunian/Desktop/python_code/simple_additive_weight/data_player_aktual.csv')
alt = df_alt['ID Pemain']
alternatif = alt.to_numpy()
print(str(alternatif))
```

```python
kriteria = ['c1','c2','c3','c4','c5','c6','c7','c8']
costbenefit = ['benefit','benefit','benefit','benefit','benefit','benefit','benefit','benefit']
kepentingan = [3,3,4,4,3,2,1,4]
alternatifkriteria = df_alt[['C1','C2','C3','C4','C5','C6','C7','C8']].to_numpy()
alternatifkriteria
```

```python
print(str(kriteria))
print(str(costbenefit))
print(str(kepentingan))
print(str(alternatifkriteria))
```

## Memasukan Rumus SAW

```python
pembagi = []
for i in range(len(kriteria)):
    pembagi.append(0)
    for j in range(len(alternatif)):
        if costbenefit[i] == 'cost':
            if j == 0:
                pembagi[i] = alternatifkriteria[j][i]
            else:
                if pembagi[i] > alternatifkriteria[j][i]:
                    pembagi[i] = alternatifkriteria[j][i]
        else: #if costbenefit[i] == 'benefit':
            if j == 0:
                pembagi[i] = alternatifkriteria[j][i]
            else:
                if pembagi[i] < alternatifkriteria[j][i]:
                    pembagi[i] = alternatifkriteria[j][i]
```

```python
print(str(pembagi))
```

## Tahap Normalisasi

```python
normalisasi = []
for i in range(len(alternatif)):
    normalisasi.append([])
    for j in range(len(kriteria)):
        normalisasi[i].append(0)
        if costbenefit[j] == 'cost':
            normalisasi[i][j] = pembagi[j] / alternatifkriteria[i][j]
        else: #if costbenefit[j] == 'benefit':
            normalisasi[i][j] = alternatifkriteria[i][j] / pembagi[j]
            
print(str(normalisasi))
```

## Hasil Skor/Nilai Pemain

```python
hasil = []
for i in range(len(alternatif)):
    hasil.append(0)
    for j in range(len(kriteria)):
        hasil[i] = hasil[i] + (normalisasi[i][j] * kepentingan[j])
        
print(str(alternatif))
```

```python
print(str(hasil))
```

## Pembentukan Kategori

```python
data_skor = ({'ID_Pemain':(alternatif),'nilai':(hasil)})
df_skor = pd.DataFrame(data_skor)
```

```python
df_n = df_skor['nilai']
n =  df_n.to_numpy()
ktr = []
def hasil():
    for i in n:
        if i < 9:
            ktr.append('pemula')
        elif i >= 9 and i <= 17:
            ktr.append('menengah')
        else :
            ktr.append('ahli')

hasil()
```

```python
df_skor.insert(2, 'kategori', ktr, True)
display(df_skor)
```
