```python
alternatif = ['lionel mesi','cristian ronaldo','muhamad salah','ponaryo astaman','robert loadowaski']
kriteria = ['IPK','jml_tanggungan','penghasilan_ortu','prestasi','jarak_rmh']
costbenefit = ['benefit','cost','benefit','benefit','cost']
kepentingan = [0.3,0.1,0.2,0.3,0.1]
alternatifkriteria = [
    [3.2,1,2,1,140],
    [3.95,3,2,4,89],
    [3.4,4,3,2,70],
    [4.0,3,4,4,120],
    [3.92,2,2,4,100]
]
```

penjumlahan

```python
pembagi = []
```

```python
for i in range(len(kriteria)):
    pembagi.append(0)
    for j in range(len(alternatif)):
        if costbenefit[i] == 'cost':
            if j == 0:
                pembagi[i] = alternatifkriteria[j][i]
            else:
                if pembagi[i] > alternatifkriteria[j][i]:
                    pembagi[i] = alternatifkriteria[j][i]
        else: #if costbenefit[i] == 'benefit':
            if j == 0:
                pembagi[i] = alternatifkriteria[j][i]
            else:
                if pembagi[i] < alternatifkriteria[j][i]:
                    pembagi[i] = alternatifkriteria[j][i]
```

```python
print(str(pembagi))
```

normalisasi

```python
normalisasi = []
```

```python
for i in range(len(alternatif)):
    normalisasi.append([])
    for j in range(len(kriteria)):
        normalisasi[i].append(0)
        if costbenefit[j] == 'cost':
            normalisasi[i][j] = pembagi[j] / alternatifkriteria[i][j]
        else: #if costbenefit[j] == 'benefit':
            normalisasi[i][j] = alternatifkriteria[i][j] / pembagi[j]
```

```python
print(str(normalisasi))
```

```python
hasil = []
```

```python
for i in range(len(alternatif)):
    hasil.append(0)
    for j in range(len(kriteria)):
        hasil[i] = hasil[i] + (normalisasi[i][j] * kepentingan[j])
#         hasil[i] = hasil[i] + normalisasi[i][j]
```

```python
print(str(alternatif))
```

```python
print(str(hasil))
```

pembentukan ranking tapi

```python
alternatifrangking = []
hasilrangking = []
```

```python
for i in range(len(alternatif)):
    hasilrangking.append(hasil[i])
    alternatifrangking.append(alternatif[i])
```

```python
for i in range(len(alternatif)):
    for j in range(len(alternatif)):
        if j > i:
            if hasilrangking[j] > hasilrangking[i]:
                tmphasil = hasilrangking[i]
                tmpalternatif = alternatifrangking[j]
                hasilrangking[i] = hasilrangking[j]
                alternatifrangking[i] = alternatifrangking[j]
                hasilrangking[j] = tmphasil
                alternatifrangking[j] = tmpalternatif
```

```python
print(str(alternatif))
```
