# Pengujian Confusion Matrics Pada Metode SAW

## Menambahkan Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
df_true = pd.read_csv('/home/adamyunian/Desktop/python_code/simple_additive_weight/data_skor_kategori_aktual.csv')
df_pred = pd.read_csv('/home/adamyunian/Desktop/python_code/simple_additive_weight/data_skor_kategori_prediksi.csv')
```

## Data Aktual

```python
df_y = df_true['kategori']
y_true = df_y.to_numpy()
y_true
```

## Data Prediksi

```python
df_yt = df_pred['kategori']
y_pred = df_yt.to_numpy()
y_pred
```

## Hasil Prediksi Metode SAW

```python
dt = confusion_matrix(y_true, y_pred, labels=['pemula','menengah','ahli'])
```

```python
tabel = pd.DataFrame(dt)
dt
```

## Menghitung Nilai Accuracy, Precision, dan Recall

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
#confution matrix
cm = confusion_matrix(y_true, y_pred)
print("CONFUSION MATRIX")
print(cm)
akurasi=classification_report(y_true, y_pred)
print("ACCURACY Simple Additive Weighting ")
print("Accuracy :", akurasi)
akurasi=accuracy_score(y_true, y_pred)
print("Tingkat : akurasi : {:.2%}".format(akurasi))
```
